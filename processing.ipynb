```python
import numpy as np
import pandas as pd
import tensorflow as tf
```

```python
tf.__version__
```

```python
#import the dataset
#df=pd.read_excel("airfoil_self_noise.xlsx")
df=pd.read_excel("airfoil_self_noise.xlsx")
df.head(1)
```

```python
x=df.drop(['ScaledSoundPressureLevel'],axis=1)
x.head(1)
```

```python
y=df[['ScaledSoundPressureLevel']]
y.head(1)
```

```python
#split the dataset
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25, random_state=2)
```

```python
from sklearn import preprocessing
max_abs_scaler = preprocessing.MinMaxScaler()
x = preprocessing.MinMaxScaler(x)
x_test = preprocessing.MinMaxScaler(x_test)
#y_train = max_abs_scaler.fit_transform(y_train)
#y_test = max_abs_scaler.fit_transform(y_test)
```

```python
x
```

```python
#building the ANN
annModel=tf.keras.models.Sequential()
```

```python
#adding the input layer and first input layer
annModel.add(tf.keras.layers.Dense(units=50, activation="relu")) #dense they are fully conncted to hidden layer neurons
#second hidden layer
annModel.add(tf.keras.layers.Dense(units=50, activation="relu")) #will be connected automatically
annModel.add(tf.keras.layers.Dense(units=50, activation="relu"))
```

```python
#create the output layer
annModel.add(tf.keras.layers.Dense(units=1)) #as we are doing output layer no activation function normally
```

```python
#compiling the ann
annModel.compile(optimizer="adam",loss="mean_squared_error")
```

```python
#build the annModel
#converting it to pandas dataframe to numpy array
x_train=np.array(x)
y_train=np.array(y)
annModel.fit(x_train,y_train,batch_size=16,epochs=1000)
```

```python
#predict the test set
from sklearn.metrics import mean_squared_error
x_test=np.array(x_test)
y_test=np.array(y_test)

y_prediction=annModel.predict(x_test)

np.set_printoptions(precision=2)

print(y_prediction)
print(mean_squared_error(y_test, y_prediction))
```

```python
annModel.save('ANN_Model') 
```
